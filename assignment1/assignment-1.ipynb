```python
# Rithvik Rajesh Matta PES2UG23CS485
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

lambda_value = 4

simulated_data = np.random.poisson(lam=lambda_value, size=1000)

plt.figure(figsize=(10, 6))
plt.hist(simulated_data, bins=range(0, max(simulated_data)+1), density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Simulated Data')

x_values = np.arange(0, max(simulated_data)+1)
poisson_pmf = stats.poisson.pmf(x_values, lambda_value)
plt.plot(x_values, poisson_pmf, 'o-', color='red', label='Theoretical Poisson PMF')

plt.title(f'Poisson Distribution Simulation (λ = {lambda_value})')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
import pandas as pd
import numpy as np
from scipy.stats import poisson

# Load the dataset (update the file path as needed)
data = pd.read_csv('customer_shopping_data.csv')

# Convert invoice_date to datetime format
data['invoice_date'] = pd.to_datetime(data['invoice_date'], dayfirst=True)

# Step 1: Calculate daily customer visits by counting unique invoices per day
daily_visits = data.groupby('invoice_date')['invoice_no'].nunique()

# Step 2: Calculate the average number of visits per day (λ)
lambda_value = daily_visits.mean()

# Step 3: Use Poisson distribution to predict probabilities for 0, 1, and 2 visits
poisson_probs = {k: poisson.pmf(k, lambda_value) for k in [0, 1, 2]}

# Step 4: Calculate actual frequencies from the data
actual_counts = daily_visits.value_counts().reindex([0, 1, 2], fill_value=0)
total_days = len(daily_visits)
actual_probs = {k: actual_counts[k] / total_days for k in [0, 1, 2]}

# Step 5: Display and compare results
results = pd.DataFrame({
    'Visits': [0, 1, 2],
    'Poisson_Probability': [poisson_probs[k] for k in [0, 1, 2]],
    'Actual_Probability': [actual_probs[k] for k in [0, 1, 2]]
})

# Output comparison
print("λ (average visits per day):", lambda_value)
print(results)
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
"""Explanation
Average Calculation (
𝜆
λ): The mean of the visits column is used as the average event rate.
Poisson Probabilities: We use the Poisson probability mass function to estimate probabilities for observing 0, 1, or 2 visits.
Actual vs. Predicted: The code compares actual probabilities (based on data frequency) with predicted probabilities to assess the model's accuracy.
Interpretation
If actual probabilities closely match predicted probabilities, it indicates that a Poisson distribution is suitable for this data. Discrepancies suggest another distribution may fit better."""
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm


df = pd.read_csv(r'healthcare_noshows_appt.csv')  # Adjust filename as needed

df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
daily_counts = df['AppointmentDay'].value_counts().sort_index()
lambda_value = daily_counts.mean()  # Average number of daily visits

x_values = np.arange(0, max(daily_counts) + 1)
poisson_probs = poisson.pmf(x_values, lambda_value)
normal_probs = norm.pdf(x_values, lambda_value, np.sqrt(lambda_value))

plt.figure(figsize=(10, 6))
plt.hist(daily_counts, bins=15, density=True, alpha=0.6, color='skyblue', label="Actual Data")
plt.plot(x_values, poisson_probs, 'o-', label="Poisson Distribution", color="red")
plt.plot(x_values, normal_probs, label="Normal Approximation", color="green")
plt.xlabel('Number of Visits')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
"""When to Use Poisson vs. Normal Distribution
Poisson: Suitable when modeling discrete events (like daily visits) occurring independently with a constant mean, especially if events are rare or the dataset has low counts.
Normal: Preferred when event counts are high or have continuous characteristics and the data distribution is roughly symmetric.
If daily counts are low or skewed, Poisson would be a better fit. However, for larger counts and more spread, a normal approximation may be reasonable."""
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r'insurance.csv')  # Adjust filename as needed
charges = df['charges']

sample_sizes = [10, 50, 100, 200]
n_samples = 1000  

plt.figure(figsize=(12, 10))

for i, size in enumerate(sample_sizes):
    means = [np.mean(np.random.choice(charges, size=size)) for _ in range(n_samples)]
    
    plt.subplot(2, 2, i + 1)
    plt.hist(means, bins=30, density=True, alpha=0.6, color='skyblue')
    plt.title(f'Sampling Distribution of the Mean (n={size})')
    plt.xlabel('Mean Charges')
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
"""Description of Results
Small Sample Size (n=10): The distribution may appear skewed or have a shape resembling the original distribution of charges, especially if the underlying data is not normal.
Medium Sample Size (n=50): The distribution begins to resemble a normal distribution but may still show some skewness.
Larger Sample Size (n=100): The sampling distribution becomes more symmetric and closer to a bell-shaped curve.
Largest Sample Size (n=200): The distribution closely resembles a normal distribution, demonstrating reduced variance and a more predictable mean.
Explanation
As the sample size increases, the sampling distribution of the mean approaches normality due to the Central Limit Theorem. This phenomenon occurs because larger samples tend to average out the variability in the data, leading to a more stable and predictable estimate of the population mean. Consequently, larger samples result in distributions with smaller standard errors, and as the sample size grows, the distribution's shape becomes more uniform and concentrated around the true population mean.

This behavior is fundamental in statistics, ensuring that researchers can reliably use sample means for inference about population parameters, particularly in hypothesis testing and confidence interval estimation.

If you need further details or have questions about specific parts of the analysis, feel free to ask!"""
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
import pandas as pd
import numpy as np

df = pd.read_csv(r'insurance.csv') 
charges = df['charges']

population_mean = charges.mean()
print(f"Population Mean of Charges: {population_mean:.2f}")

sample_size = 50
n_samples = 1000

sample_means = [np.mean(np.random.choice(charges, size=sample_size)) for _ in range(n_samples)]

sampling_distribution_mean = np.mean(sample_means)
print(f"Mean of Sampling Distribution (n={sample_size}): {sampling_distribution_mean:.2f}")
difference = abs(population_mean - sampling_distribution_mean)
print(f"Difference between Population Mean and Sampling Distribution Mean: {difference:.2f}")
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
"""Observations
Population Mean: This represents the true mean of the charges variable for the entire dataset.
Mean of Sampling Distribution: The calculated mean of the 1000 sample means should be very close to the population mean.
Difference: You can expect a small difference between these two means, which supports the Central Limit Theorem (CLT).
Explanation of Results
The CLT states that the distribution of the sample means will be approximately normally distributed if the sample size is sufficiently large (typically 
𝑛
≥
30
n≥30), regardless of the shape of the population distribution. In this case:

As you calculate the means of 1000 samples of size 50, the average of these means will converge to the population mean.
The small difference observed validates the CLT, demonstrating that with larger sample sizes, the sample mean is a reliable estimator of the population mean.
Conclusion
This exercise illustrates the robustness of the sample mean as an estimator for the population mean and underscores the importance of the Central Limit Theorem in statistical inference. If you have further questions or need clarification on any part of this analysis, feel free to ask!"""
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
import pandas as pd
import numpy as np

df = pd.read_csv(r'insurance.csv')  
charges = df['charges']

sample_sizes = [30, 60, 120]
n_samples = 1000

std_devs = {}

for size in sample_sizes:
    sample_means = [np.mean(np.random.choice(charges, size=size)) for _ in range(n_samples)]
    std_dev = np.std(sample_means)
    std_devs[size] = std_dev
    print(f"Standard Deviation of Sampling Distribution (n={size}): {std_dev:.2f}")

print("\nSummary of Standard Deviations:")
for size, std_dev in std_devs.items():
    print(f"Sample Size {size}: Standard Deviation = {std_dev:.2f}")
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
"""Observations
Standard Deviation of Sampling Distributions: For each sample size, the code calculates and prints the standard deviation of the sampling distribution of the sample mean.
Expected Results:
You should observe that the standard deviation decreases as the sample size increases. This reflects the concept of the standard error of the mean, which is calculated as 
𝜎
/
𝑛
σ/ 
n
​
  (where 
𝜎
σ is the population standard deviation and 
𝑛
n is the sample size).
Explanation of Results
Effect of Sample Size on Standard Deviation:

As the sample size increases, the variability of the sample means decreases. This means that larger samples provide more reliable estimates of the population mean.
The standard error (the standard deviation of the sample means) decreases with increasing sample size due to the averaging effect, where random variations in smaller samples are mitigated.
Central Limit Theorem:

The findings support the Central Limit Theorem, which asserts that the distribution of sample means approaches normality as the sample size increases, leading to a smaller standard deviation of the sampling distribution.
Conclusion
This analysis illustrates how sample size impacts the standard deviation of the sampling distribution of the mean. Larger samples yield more precise estimates of the population mean due to reduced variability, highlighting the importance of sample size in statistical analysis.
"""
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'bmi.csv') 
bmi_values = df['Bmi']

sample_sizes = [10, 30, 100]
n_samples = 1000

plt.figure(figsize=(12, 10))

for i, size in enumerate(sample_sizes):
    sample_means = [np.mean(np.random.choice(bmi_values, size=size)) for _ in range(n_samples)]
    
    plt.subplot(2, 2, i + 1)
    sns.histplot(sample_means, bins=30, kde=True, color='skyblue', stat="density")
    plt.axvline(np.mean(sample_means), color='red', linestyle='dashed', linewidth=1)
    plt.axvline(np.mean(bmi_values), color='green', linestyle='dashed', linewidth=1, label='Population Mean')
    plt.title(f'Sampling Distribution of the Mean (Sample Size = {size})')
    plt.xlabel('Sample Mean BMI')
    plt.ylabel('Density')
    plt.legend()

plt.subplot(2, 2, 4)
sns.histplot(bmi_values, bins=30, kde=True, color='orange', stat="density")
plt.title('Population Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
# Rithvik Rajesh Matta PES2UG23CS485
"""Explanation of the Results
Population Distribution: The BMI data may exhibit a skewed distribution, often tending towards a right skew (more individuals with lower BMI and fewer with higher values).

Sampling Distributions:

Small Sample Size (n=10): The sampling distribution of the sample mean may still show skewness similar to the original distribution of BMI.
Medium Sample Size (n=30): The distribution begins to flatten out, showing signs of normality.
Large Sample Size (n=100): The sampling distribution of the sample mean should resemble a normal distribution, demonstrating that the mean values are clustering around the population mean.
Conclusion
The Central Limit Theorem states that as sample sizes increase, the distribution of sample means will tend to be normal, regardless of the population's distribution. Even when the original data (like BMI) is skewed, larger sample sizes lead to a sampling distribution of the mean that approaches normality. This validates the CLT and shows its robustness across different types of distributions."""
```
