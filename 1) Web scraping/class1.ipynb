```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
import time
import time
from bs4 import BeautifulSoup
 
 
options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)
driver = webdriver.Chrome(options=options)
driver.get("https://patentscope.wipo.int/search/en/search.jsf")
 
 
wait = WebDriverWait(driver, 20)
input_field = wait.until(EC.presence_of_element_located((By.ID, 'simpleSearchForm:fpSearch:input')))
input_field.clear()
 
input_field.send_keys('hi')
print("Search text entered successfully.")
 
submit_button = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "b-button-medium.primary.js-default-button")))
submit_button.click()
print("Submit button clicked successfully.")
 
 
time.sleep(5)  
page_html = driver.page_source
soup = BeautifulSoup(page_html, 'html.parser')
 
span_elements = soup.find_all('span', class_='trans-section needTranslation-biblio')
for span in span_elements:
    print(span.get_text(strip=True))
```
